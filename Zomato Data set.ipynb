## Observations
- when the rating is from 4.5 to 4.9 then it is rated excellent
- when the rating is from 4.0 to 4.4 then it is rated Very Good
- when the rating is from 3.5 to 3.9 then it is rated Good
- when the rating is from 3.0 to 3.4 then it is rated Average
- when the rating is from 0.0 to 2.9 then it is rated Poor
- most of the people have not yet rated

# Find the top 10 cusines

```python
final_zomato_data.columns
```

```python
cusines_counts = final_zomato_data.groupby("Cuisines").size().reset_index()
cusines_counts
```

```python
sorted_cusines = cusines_counts.sort_values(by=0,ascending=False)
sorted_cusines
```

```python

# Get the bar names
bar_names = sorted_cusines["Cuisines"]

# Get the bar heights
bar_heights = sorted_cusines[0]

# Create a bar chart of the bar heights
plt.bar(bar_names[:10], bar_heights[:10])
plt.xlabel("Cusines")
plt.ylabel("Counts")
plt.xticks(rotation=90)
plt.title("Preferrable Cusines")
plt.show()
```
